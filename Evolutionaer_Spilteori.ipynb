# Projekt 1: Gittermodel — Rumlig Spilteori

I de tidligere opgaver har vi antaget, at ethvert dyr kan møde ethvert andet dyr (en "velblandet" population). I virkeligheden bor dyr i et landskab, og de møder primært deres naboer.

I gittermodellen placerer vi hvert dyr på et felt i et $L \times L$ gitter. Hvert dyr spiller spillet mod sine 4 (eller 8) nærmeste naboer og kan kopiere den bedst stillede nabos strategi med en sandsynlighed proportional med gevinstforskellen.

Funktionen er `evspil.simuler_gitter(L, andel_orn, V, C, n_skridt, naboer, seed)`.

## 1.1 Første simulering

Vi starter med et lille gitter, så I kan se mønstrene tydeligt. Brug `plot_gitter_oversigt` til at få både snapshots og tidsserie i samme figur.

```python
import time

res = evspil.simuler_gitter(L=40, andel_orn=0.5, V=10, C=20,
                             n_skridt=80, naboer=4, seed=1)

evspil.plot_gitter_oversigt(res, V=10, C=20,
                            titel='Gittermodel: V=10, C=20, L=40')
plt.show()
print(f'Slut-andel ørne: {res["andel_orn"][-1]:.3f}')
print(f'Velblandet ESS: V/C = {evspil.ess_orn_due(10, 20):.3f}')
```

Bemærk hvordan ørne (rød) og duer (blå) ikke blandes tilfældigt — de samles i klynger. Duer kan overleve i midten af en due-klynge, fordi de der kun møder andre duer.

## 1.2 Hurtig kontra langsom tilstand

Lad os prøve at slå hurtig tilstand fra og se forskellen. Begge skal give samme dynamik (men med forskellig hastighed).

```python
# Hurtig
evspil.HURTIG_TILSTAND = True
t0 = time.time()
res_h = evspil.simuler_gitter(L=40, n_skridt=80, V=10, C=20, seed=42)
t_h = time.time() - t0

# Langsom
evspil.HURTIG_TILSTAND = False
t0 = time.time()
res_l = evspil.simuler_gitter(L=40, n_skridt=80, V=10, C=20, seed=42)
t_l = time.time() - t0

# Slå hurtig tilstand til igen
evspil.HURTIG_TILSTAND = True

print(f'L=40, 80 skridt:')
print(f'  Hurtig:  {t_h:.3f}s')
print(f'  Langsom: {t_l:.3f}s')
print(f'  Hurtigere: {t_l/t_h:.0f}x')

fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(res_h['andel_orn'], label='hurtig tilstand', linewidth=2)
ax.plot(res_l['andel_orn'], label='langsom tilstand', linewidth=2, linestyle='--')
ax.set_xlabel('Tidsskridt'); ax.set_ylabel('Andel ørne')
ax.legend(); ax.grid(True, alpha=0.3)
ax.set_title('Begge tilstande giver samme dynamik')
plt.show()
```

# Projekt 2: Sæsonøkosystem

**Tidsforbrug: ca. 40 minutter**

I virkelige økosystemer skifter ressourcerne med årstiderne. Vi udvider individ-simuleringen, så $V$ kan være en funktion af tiden $V(t)$, og observerer hvordan både populationsstørrelsen og strategi-andelen reagerer.

Funktionen er `evspil.simuler_individer_dynamisk(N, andel_orn, V_funktion, C, n_skridt, ...)` og resultatet plottes med `evspil.plot_dynamisk_simulering(resultat)`.

## 2.1 Stabile sæsoner

Vi modellerer årstiderne med en sinusfunktion: $V(t) = V_0 + A \sin(2\pi t / T)$, hvor $T$ er periodelængden. Først prøver vi en stabil situation.

```python
V0 = 12       # gennemsnitlig ressourcemængde
A = 6         # amplitude (sæsonudsvingets størrelse)
T = 60        # periodelængde ("år" i tidsskridt)
C = 16        # konstant kampomkostning

def V_sæson(t):
    return V0 + A * np.sin(2 * np.pi * t / T)

res = evspil.simuler_individer_dynamisk(
    N=400, andel_orn=0.5,
    V_funktion=V_sæson, C=C,
    n_skridt=400, doedsrate=0.05, seed=1)

evspil.plot_dynamisk_simulering(res, vis_V=True,
    titel='Stabile sæsoner: V(t) = 12 + 6·sin(2πt/60), C = 16')
plt.show()
```

Andelen af ørne følger den øjeblikkelige $p^*(t) = V(t)/C$ tæt — om sommeren vokser ørne-andelen, om vinteren falder den. Populationsstørrelsen følger med.

## 2.2 Hvor vilde må sæsonerne være?

Vi sammenligner forskellige amplituder $A$. Hvor stort kan udsvinget blive før arten kollapser?

```python
C = 16
fig, axes = plt.subplots(2, 2, figsize=(12, 7), sharex=True)

for ax, A in zip(axes.flat, [3, 6, 9, 11]):
    def V_t(t, A=A):
        return 12 + A * np.sin(2 * np.pi * t / 60)
    res = evspil.simuler_individer_dynamisk(
        N=400, andel_orn=0.5, V_funktion=V_t, C=C,
        n_skridt=400, doedsrate=0.05, seed=1)
    ax.plot(res['befolkning'], color='#1b7837', linewidth=2, label='population')
    ax2 = ax.twinx()
    ax2.plot(res['andel_orn'], color='#c1272d', linewidth=2, label='andel ørne')
    ax2.set_ylim(-0.05, 1.05)
    ax.set_title(f'Amplitude A = {A}')
    ax.set_ylabel('Population', color='#1b7837')
    ax2.set_ylabel('Andel ørne', color='#c1272d')
    ax.grid(True, alpha=0.3)
fig.suptitle('Voksende sæsonudsving (C=16, V₀=12)', fontsize=13)
fig.tight_layout()
plt.show()
```

Ved store amplituder kommer $V$ tæt på 0 om vinteren, og populationen kan kollapse helt. Ved moderate amplituder følger systemet sæsonerne pænt.

## 2.3 Et tørke-år

Forestil jer en sæson hvor *et helt år* har meget lav $V$ — fx pga.\ klimaforandringer.

```python
def V_med_torke(t, torke_start=120, torke_slut=180, torke_V=4):
    base = 12 + 6 * np.sin(2 * np.pi * t / 60)
    if torke_start <= t < torke_slut:
        return min(base, torke_V)
    return base

res = evspil.simuler_individer_dynamisk(
    N=400, andel_orn=0.5, V_funktion=V_med_torke, C=16,
    n_skridt=400, doedsrate=0.05, seed=1)

evspil.plot_dynamisk_simulering(res, vis_V=True,
    titel='Et tørke-år midt i forløbet (t = 120–180)')
plt.show()

print(f'Population før tørken (t=100): {res["befolkning"][100]}')
print(f'Min population under tørken: {res["befolkning"][120:180].min()}')
print(f'Population efter tørken (t=300): {res["befolkning"][300]}')
```

## 2.4 Hvor barsk en tørke kan arten overleve?

Vi tester forskellige tørke-niveauer og hvor mange kørsler der overlever ud af 5 (med forskellige seeds).

```python
torke_V_range = [1, 2, 3, 4, 6, 8]
n_seeds = 5
overlevelses_rate = []
min_pop = []

for V_torke in torke_V_range:
    overlevet = 0
    pop_min = []
    for seed in range(n_seeds):
        def V_t(t, V_torke=V_torke):
            base = 12 + 6 * np.sin(2*np.pi*t/60)
            return min(base, V_torke) if 120 <= t < 180 else base
        res = evspil.simuler_individer_dynamisk(
            N=400, andel_orn=0.5, V_funktion=V_t, C=16,
            n_skridt=400, doedsrate=0.05, seed=seed)
        if res['befolkning'][-1] > 10:  # arten overlevede
            overlevet += 1
        pop_min.append(res['befolkning'][120:200].min())
    overlevelses_rate.append(overlevet / n_seeds)
    min_pop.append(np.mean(pop_min))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(torke_V_range, overlevelses_rate, 'o-', color='#c1272d',
             linewidth=2, markersize=10)
axes[0].set_xlabel('V under tørke')
axes[0].set_ylabel('Overlevelses-rate (af 5 kørsler)')
axes[0].set_title('Hvornår overlever arten tørken?')
axes[0].set_ylim(-0.05, 1.05)
axes[0].grid(True, alpha=0.3)

axes[1].bar(range(len(torke_V_range)), min_pop, color='#1b7837', alpha=0.7)
axes[1].set_xticks(range(len(torke_V_range)))
axes[1].set_xticklabels([f'V={v}' for v in torke_V_range])
axes[1].set_ylabel('Mindste population under tørken (gennemsnit)')
axes[1].set_title('Hvor lav når populationen?')
axes[1].grid(True, alpha=0.3)
fig.tight_layout()
plt.show()
```

## 2.5 Lange vs.\ korte sæsoner

Hvad sker der hvis sæsonerne er meget korte (måske dag-nat-cykler) eller meget lange (årtier)?

```python
fig, axes = plt.subplots(3, 1, figsize=(11, 8), sharex=True)

for ax, T in zip(axes, [20, 60, 200]):
    def V_t(t, T=T):
        return 12 + 6 * np.sin(2 * np.pi * t / T)
    res = evspil.simuler_individer_dynamisk(
        N=400, andel_orn=0.5, V_funktion=V_t, C=16,
        n_skridt=600, doedsrate=0.05, seed=1)
    ax.plot(res['andel_orn'], color='#c1272d', linewidth=2, label='andel ørne')
    p_star = res['V_historik'] / 16
    p_star = np.minimum(p_star, 1.0)
    ax.plot(p_star, color='black', linestyle='--', alpha=0.5, label='p*(t) = V(t)/C')
    ax.set_title(f'Periodelængde T = {T}')
    ax.set_ylabel('Andel ørne')
    ax.set_ylim(-0.05, 1.05)
    ax.legend(loc='upper right'); ax.grid(True, alpha=0.3)
axes[-1].set_xlabel('Tidsskridt')
fig.suptitle('Hurtige vs. langsomme sæsoner', fontsize=13)
fig.tight_layout()
plt.show()
```

**Konklusion fra projekt 2:**
- Ved langsomme sæsoner følger andelen af ørne $p^*(t)$ præcist
- Ved hurtige sæsoner kan populationen ikke nå at adaptere; andelen ligger fast omkring gennemsnittet
- Korte ekstreme tørker er farligere end milde langsomme variationer, fordi populationen kan blive for lille til at komme sig

# Projekt 3: Miljøskift og Adaptation

**Tidsforbrug: ca. 40 minutter**

Forestil jer en art, der har levet i ligevægt i lang tid, og pludselig ændrer miljøet sig **permanent**. Fx kan kampens omkostning $C$ stige fordi farligere rovdyr er kommet til, eller fordi sygdomme er blevet alvorligere.

Spørgsmålet er: kan arten tilpasse sig hurtigt nok? Mutation kan hjælpe — men hvor meget mutation er optimalt?

I dette projekt undersøger vi:
- Hvad der sker uden mutation (populationen sidder fast)
- Hvordan mutation muliggør adaptation
- Den optimale mutationsrate (hverken for lav eller for høj)
- Hvad der sker i et kronisk ustabilt miljø

## 3.1 Stabilt miljø — definér før og efter

Vi bestemmer to forskellige værdier af $C$ — før og efter skiftet — og beregner ESS for hver.

```python
V = 10
C_før = 14    # før skiftet — moderat kamp
C_efter = 28  # efter skiftet — kamp er nu meget farligere
T_skift = 200

fig, ax = plt.subplots(figsize=(9, 4))
p_før = V / C_før
p_efter = V / C_efter
ax.bar(['Før (C=14)', 'Efter (C=28)'], [p_før, p_efter],
       color=['#0072b2', '#c1272d'], alpha=0.8)
ax.set_ylabel('Stabil andel ørne (p* = V/C)')
ax.set_ylim(0, 1)
for i, p in enumerate([p_før, p_efter]):
    ax.text(i, p + 0.02, f'{p:.3f}', ha='center', fontweight='bold')
ax.set_title('Hvor skal andelen af ørne hen?')
ax.grid(True, alpha=0.3, axis='y')
plt.show()

print(f'Skift skal være: {p_før - p_efter:+.3f} i ørne-andel')
```

## 3.2 Skift uden mutation

Vi simulerer skiftet ved $t = 200$. Med mutationsrate $\mu = 0$ kan systemet kun ændre proportionerne af de eksisterende strategier.

```python
def C_skift(t):
    return C_før if t < T_skift else C_efter

res = evspil.simuler_individer_dynamisk(
    N=600, andel_orn=V/C_før, V=V, C_funktion=C_skift,
    n_skridt=500, doedsrate=0.05, mutationsrate=0.0, seed=1)

evspil.plot_dynamisk_simulering(res, vis_V=False, vis_C=True,
    titel='Miljøskift ved t=200, ingen mutation')
plt.show()

print(f'Andel ørne før skiftet (t=180-200): {res["andel_orn"][180:200].mean():.3f}')
print(f'Andel ørne efter skiftet (t=400-500): {res["andel_orn"][400:].mean():.3f}')
print(f'Forventet ny ESS: {V/C_efter:.3f}')
print(f'Slut-population: {res["befolkning"][-1]}')
```

## 3.3 Forskellige mutationsrater

Hvad sker der med forskellige mutationsrater? Vi prøver fire værdier.

```python
fig, axes = plt.subplots(2, 1, figsize=(11, 6.5), sharex=True)
farver = plt.cm.plasma(np.linspace(0.1, 0.85, 4))

for mu, farve in zip([0.0, 0.01, 0.05, 0.20], farver):
    res = evspil.simuler_individer_dynamisk(
        N=600, andel_orn=V/C_før, V=V, C_funktion=C_skift,
        n_skridt=500, doedsrate=0.05, mutationsrate=mu, seed=1)
    axes[0].plot(res['andel_orn'], color=farve, linewidth=2, label=f'μ = {mu}')
    axes[1].plot(res['befolkning'], color=farve, linewidth=2, label=f'μ = {mu}')

axes[0].axvline(T_skift, color='black', linestyle=':', alpha=0.6,
                label=f'skift ved t={T_skift}')
axes[0].axhline(V/C_efter, color='red', linestyle='--', alpha=0.5,
                label=f'ny ESS = {V/C_efter:.2f}')
axes[0].set_ylabel('Andel ørne')
axes[0].legend(loc='upper right', ncol=2); axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(-0.05, 1.05)
axes[1].axvline(T_skift, color='black', linestyle=':', alpha=0.6)
axes[1].set_ylabel('Population'); axes[1].set_xlabel('Tidsskridt')
axes[1].grid(True, alpha=0.3); axes[1].legend(loc='upper right')
fig.suptitle('Effekt af mutationsraten', fontsize=13)
fig.tight_layout()
plt.show()
```

Bemærk:
- $\mu = 0$: ingen adaptation, populationen kan kollapse
- $\mu = 0.01$: langsom adaptation
- $\mu = 0.05$: hurtig adaptation, lille bias væk fra ESS
- $\mu = 0.20$: mutation overdøver selektion — andelen drives mod 50/50

## 3.4 Hvor hurtigt tilpasser arten sig?

Vi måler **adaptationstid** — antal skridt fra skiftet, til populationen er kommet inden for 0.05 af den nye ESS.

```python
mu_range = np.array([0.005, 0.01, 0.02, 0.05, 0.10, 0.20, 0.30])
adaptationstid = []
slut_pop = []
n_seeds = 4

for mu in mu_range:
    tider = []
    pops = []
    for seed in range(n_seeds):
        res = evspil.simuler_individer_dynamisk(
            N=600, andel_orn=V/C_før, V=V, C_funktion=C_skift,
            n_skridt=600, doedsrate=0.05, mutationsrate=mu, seed=seed)
        # Find første skridt efter T_skift hvor andelen er inden for 0.05 af ny ESS
        target = V/C_efter
        t_recovered = None
        for t in range(T_skift, len(res['andel_orn'])):
            # Glat med vindue for at undgå støj
            if t < len(res['andel_orn']) - 10:
                if abs(res['andel_orn'][t:t+10].mean() - target) < 0.05:
                    t_recovered = t - T_skift
                    break
        if t_recovered is not None:
            tider.append(t_recovered)
        pops.append(res['befolkning'][-1])
    adaptationstid.append(np.mean(tider) if tider else np.nan)
    slut_pop.append(np.mean(pops))

fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))
axes[0].plot(mu_range, adaptationstid, 'o-', color='#c1272d', linewidth=2, markersize=10)
axes[0].set_xscale('log')
axes[0].set_xlabel('Mutationsrate μ')
axes[0].set_ylabel('Adaptationstid (tidsskridt)')
axes[0].set_title('Hvor hurtigt finder populationen den nye ESS?')
axes[0].grid(True, alpha=0.3, which='both')

axes[1].plot(mu_range, slut_pop, 's-', color='#1b7837', linewidth=2, markersize=10)
axes[1].set_xscale('log')
axes[1].set_xlabel('Mutationsrate μ')
axes[1].set_ylabel('Slut-population')
axes[1].set_title('Hvor stor er populationen til sidst?')
axes[1].grid(True, alpha=0.3, which='both')
fig.tight_layout()
plt.show()
```

Adaptationstid falder med højere $\mu$ — men ved meget høj $\mu$ får populationen aldrig fat i ESS, så slut-populationen kan også lide.

## 3.5 Kronisk ustabilt miljø

Hvad sker der hvis miljøet skifter cyklisk mellem to værdier?

```python
fig, axes = plt.subplots(2, 1, figsize=(11, 6), sharex=True)

for ax, periode in zip(axes, [120, 30]):
    def C_cyklisk(t, periode=periode):
        return 14 if (t // periode) % 2 == 0 else 28
    res = evspil.simuler_individer_dynamisk(
        N=500, andel_orn=0.5, V=10, C_funktion=C_cyklisk,
        n_skridt=600, doedsrate=0.05, mutationsrate=0.05, seed=1)
    ax.plot(res['andel_orn'], color='#c1272d', linewidth=2, label='andel ørne')
    ax.plot(np.minimum(res['V_historik']/np.array(res['C_historik']), 1.0),
            color='black', linestyle='--', alpha=0.5, label='p*(t)')
    ax.set_title(f'Cyklus-længde = {periode} skridt')
    ax.set_ylabel('Andel ørne')
    ax.set_ylim(-0.05, 1.05)
    ax.legend(loc='upper right'); ax.grid(True, alpha=0.3)
axes[-1].set_xlabel('Tidsskridt')
fig.suptitle('Langsomme vs. hurtige miljøskift', fontsize=13)
fig.tight_layout()
plt.show()
```

Ved lange cyklusser kan populationen tilpasse sig hver gang. Ved korte cyklusser hænger populationen ikke med — andelen sidder fast et sted i midten.

## 3.6 Sammenligning og konklusion

Vi laver et opsummerende heatmap: gennemsnitlig andel ørne efter skiftet, som funktion af både mutationsrate og skift-størrelse.

```python
mu_range = [0.0, 0.01, 0.05, 0.1, 0.2]
C_efter_range = [16, 20, 24, 28, 32, 40]

matrix = np.zeros((len(mu_range), len(C_efter_range)))
for i, mu in enumerate(mu_range):
    for j, C_e in enumerate(C_efter_range):
        def C_func(t, C_e=C_e):
            return 14 if t < 200 else C_e
        res = evspil.simuler_individer_dynamisk(
            N=400, andel_orn=10/14, V=10, C_funktion=C_func,
            n_skridt=500, doedsrate=0.05, mutationsrate=mu, seed=1)
        matrix[i, j] = res['andel_orn'][400:].mean()

fig, ax = plt.subplots(figsize=(9, 5))
im = ax.imshow(matrix, cmap='RdBu_r', aspect='auto', origin='lower',
               vmin=0, vmax=1)
ax.set_xticks(range(len(C_efter_range)))
ax.set_xticklabels([f'C={c}\n(p*={10/c:.2f})' for c in C_efter_range])
ax.set_yticks(range(len(mu_range)))
ax.set_yticklabels([f'μ={mu}' for mu in mu_range])
ax.set_xlabel('Ny kampomkostning C')
ax.set_ylabel('Mutationsrate μ')
ax.set_title('Andel ørne efter miljøskift (gennemsnit af t=400–500)')
for i in range(len(mu_range)):
    for j in range(len(C_efter_range)):
        ax.text(j, i, f'{matrix[i,j]:.2f}', ha='center', va='center',
                color='white' if abs(matrix[i,j]-0.5) > 0.25 else 'black')
plt.colorbar(im, ax=ax, label='Andel ørne')
fig.tight_layout()
plt.show()
```

**Konklusion fra projekt 3:**
- Uden mutation kan populationen ikke flytte sig fra én ESS til en anden
- Lav mutation (1–5%) giver langsom men sikker adaptation
- Høj mutation (>20%) gør populationen ude af stand til at finde ESS — selektion overdøves
- I et kronisk ustabilt miljø afhænger den optimale strategi af, hvor hurtigt miljøet skifter ift.\ adaptationshastigheden

Disse mekanismer er centrale for at forstå hvorfor evolution er **muligt** — uden mutation ville arter være fastlåst i deres første ligevægt.
